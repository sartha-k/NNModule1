```python
#Create Model Class
import torch
import torch.nn as nn # contains neural network components ex- nn.Linear nn.Conv2d nn.Sogmoid nn.Relu nn.CrossEntropyLoss

class model(nn.Module): # All pytorch model must inherit from this class so that pytorch can manage your network
  def __init__(self,num_feature): #this run when model is created
    super().__init__() # initialize the nn.module parent class
    self.linear = nn.Linear(num_feature,1) #  input size =  num_feature and output size = 1 # weight shape = (1, num_feature) is how it's STORED
# but the actual matmul is (batch_size, num_feature) @ (num_feature, 1) → (batch_size, 1)) bias  shape = (1)
    self.sigmoid = nn.Sigmoid() #Because we are using binary classification
  def forward(self,feature): # forwardpropagation function
    out = self.linear(feature) # feature shape = (batch_size, num_feature)
# weight stored as (1, num_feature), transposed to (num_feature, 1) during matmul
# result = (batch_size, 1)
    out = self.sigmoid(out)  # Sigmoid function is used  for activation

    return out # return the final prediction


```

```python
#create dataset
feature = torch.rand(10,5)

#create model
model = model(feature.shape[1])

#call model for forward pass
model(feature)
```
